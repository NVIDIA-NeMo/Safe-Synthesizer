```python
def embed_text(df: pd.DataFrame, embedder) -> pd.DataFrame:
    """Takes a dataframe of text fields, finds the embeddings for each
    and then averages the embeddings into one embedding and returns a dataframe with just that
    """
    embeddings = {}
    # for col in df.columns:
    #     print(f"Embedding column: {col}")
    #     data = df[col].to_list()
    #     data = [str(r) for r in data]
    #     embeddings[col] = embedder.encode(data, show_progress_bar=False, convert_to_numpy=True)
    #     print(f"Embeddings for column {col}: {embeddings[col].shape}")

    for j in range(len(df.columns)):
        col = df.columns[j]
        embd = np.array([[0.1, 0.2, 0.3]] * 15) * j
        embeddings[col] = embd
        print(f"Embeddings for column {col}: {embd}")
    print("done embedding")

    stacked = np.stack([embeddings[col] for col in df.columns], axis=0)  # shape: (n_cols, n_rows, embed_dim)
    avg_embeddings = np.mean(stacked, axis=0)  # shape: (n_rows, embed_dim)

    return pd.DataFrame({"embedding": list(avg_embeddings)})
```
